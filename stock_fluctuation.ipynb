```python
start_date = datetime.datetime(2018,1,1)     #開始日期
end_date= datetime.datetime(2018,12,31)      #結束日期
```

```python
df_2327 = pdr.DataReader('2327.TW', 'yahoo', start=start_date ,end=end_date)   # 台灣國巨
df_2327.head()
```

```python
df_2327['Date']=df_2327.index             # index 複製到另一欄
df_2327 = df_2327.reset_index(drop=True)  # index變成 0~N
df_2327.head()
```

```python
##增欄位:fluctuation幅度 tag漲跌平
df_2327['fluctuation']=0
df_2327['tag']='--'
df_2327.head()
```

```python
###計算漲跌
###上漲>3%標示為up 下跌標示為down
for index, row in df_2327.iterrows():
    try:
        margin =(float(df_2327.loc[index,'High']) - float(df_2327.loc[index-1,'Close']) )/ float(df_2327.loc[index-1,'Close'])
        df_2327.loc[index,'fluctuation']=margin
        if margin >0.03:
            df_2327.loc[index,'tag']='up'
        elif margin<-0.03:
            df_2327.loc[index,'tag']='down'
    except:
        continue
```

```python
###將上漲的資料取出
up_fliter=(df_2327['tag']=='up')
up_2327=df_2327[up_fliter]
up_2327.head()
```

```python
###將下跌的資料取出
down_fliter=(df_2327['tag']=='down')
down_2327=df_2327[down_fliter]
down_2327
```

```python
###將日期做規格處理
upDate=up_2327["Date"].astype(str)
downDate=down_2327["Date"].astype(str)
downDate
```

```python
###將日期做規格處理
for key,value in downDate.items():
    downDate[key]=datetime.datetime.strptime(value,"%Y-%m-%d")
    print(downDate[key])
```

```python
###將日期做規格處理
for key,value in upDate.items():
    upDate[key]=datetime.datetime.strptime(value,"%Y-%m-%d")
    print(upDate[key])
```

```python
###讀取論壇資料
df_forum = pd.read_csv("bda2019_dataset/forum2.csv",encoding="utf-8")
df_forum[['post_time','title','content']]
```

```python
###開始做兩種資料表中的時間比對
##資料時間規格處理
forumDate=df_forum['post_time'].astype(str)
for i in range(len(forumDate)):
    try:
        forumDate[i]=datetime.datetime.strptime(forumDate[i],"%Y-%m-%d %H:%M:%S").replace(hour=0, minute=0,second=0)
    except ValueError:
        forumDate[i]=datetime.datetime.strptime(forumDate[i],"%Y/%m/%d %H:%M:%S").replace(hour=0, minute=0,second=0)
forumDate.head()
```

```python
###文章處理
cutlist = "「」<>/:：;；,、，。？＂’，.。！？?,.｢\"\'\\\n\r《》“”!@#$%^&*()0123456789（）abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
down_title_list=[]
down_content_list=[]
up_title_list=[]
up_content_list=[]
```

```python
###符號處裡 (未完全
def deDot(line):
        sentence = ""
        for word in line:
            if word not in cutlist: #如果文字不是標點符號，就把字加到句子中
                sentence += word
        return(sentence)
```

```python
###比對時間，若一樣則將標題和內容存進list
for key,value in downDate.items():
    for i in range(len(forumDate)):
        if(not (downDate[key].strftime("%Y")==forumDate[i].strftime("%Y"))):
            continue
        if(not (downDate[key].strftime("%m")==forumDate[i].strftime("%m"))):
            continue
        if(downDate[key].strftime("%d")==forumDate[i].strftime("%d")):
            down_title_list.append(deDot(df_forum.loc[i,'title']))
            down_content_list.append(deDot(df_forum.loc[i,'content']))
```

```python
for key,value in upDate.items():
    for i in range(len(forumDate)):
        if(not (upDate[key].strftime("%Y")==forumDate[i].strftime("%Y"))):
            continue
        if(not (upDate[key].strftime("%m")==forumDate[i].strftime("%m"))):
            continue
        if(upDate[key].strftime("%d")==forumDate[i].strftime("%d")):
            up_title_list.append(deDot(df_forum.loc[i,'title']))
            up_content_list.append(deDot(df_forum.loc[i,'content']))
```
